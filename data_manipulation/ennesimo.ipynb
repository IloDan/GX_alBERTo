```python
import pandas as pd

gxbert = pd.read_csv('../dataset/GX-BERT_tobuild37.csv', sep=',')

lizio = pd.read_csv('../dataset/Join_with_Lizio.csv', sep=',')
```

```python
gxbert
```

```python
lizio
```

```python
merged= pd.merge(gxbert, lizio, on='gene_name', how='inner')
merged
```

```python
#prendi i geneid in gxbert che non sono in merged
gxbert_not_in_merged = gxbert[~gxbert.gene_id.isin(merged.gene_id_x)]
gxbert_not_in_merged
```

```python
ensemble = pd.read_csv('../dataset/ens37_mancanti.txt', sep='\t')
ensemble
```

```python
#merge con ensemble
merged_ensemble= pd.merge(gxbert_not_in_merged, ensemble, on='gene_id', how='inner')
merged_ensemble
```

```python
df4 = merged_ensemble.loc[merged_ensemble.groupby("gene_id")["Transcript length (including UTRs and CDS)"].idxmax()]
df4
```

```python

order=df4.columns.tolist()
df6=df4
```

```python
#aggiungi chr di fianco a chromosome_name
df6['chromosome_name'] = df6['chromosome_name'].apply(lambda x: 'chr'+str(x))
df6
```

```python
merged=merged[order]
merged
```

```python
#concatena i due dataframe
result = pd.concat([merged, df6])
result
```

```python
result.to_csv('../dataset/FINITOPD.csv', index=False)
```
